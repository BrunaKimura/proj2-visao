Para facilitar a experimentação:

- O Scikit-Image já tem uma função que faz o filtro de Wiener: ``skimage.restoration.wiener``

- Segue código para fazer um *kernel* de *motion blur* dado o tamanho do kernel, o comprimento do segmento de reta, e o ângulo.

```python
from skimage.restoration import wiener

def make_psf(shape, length, angle):
    psf = np.zeros(shape)
    rows, cols = shape
    center = (cols//2, rows//2)
    psf[center[1], center[0]:(center[0]+length)] = 1.0
    M = cv2.getRotationMatrix2D(center, angle, 1)
    psf = cv2.warpAffine(psf, M, (cols, rows))
    psf = psf / psf.sum()
    return psf
```

## Método aula 6


Para corrigir o motion blur foram feitas uma mesma técnica de duas formas diferentes.
A primeira técnica foi utilizar o código para gerar o filtro de Wiener disponibilizado pelo professor. Para tanto criamos a função wiener1 que simula esse filtro.

```python
def wiener1(G, H, K):
    W = H.conj() / (np.abs(H)**2 + K)
    F = W*G
    return F
```

```python
# Chuta o K.
K = 0.001

l=20
ang=-10

# Chuta o h
h = make_psf((img_blur.shape[0],img_blur.shape[1]), l, ang)
H = np.fft.fft2(h)

G = np.fft.fft2(img_blur)
img_test_blur_recon_wiener_freq = wiener1(G, H, K)
img_blur_recon = np.fft.fftshift(np.fft.ifft2(img_test_blur_recon_wiener_freq).real)

plt.figure(figsize=(8, 8))
plt.imshow(np.clip(img_blur_recon, 0, 1), cmap='gray');
```

```python
#redimencionando a janela de exibição da placa
rows, cols = 400, 1000

#achamos os pontos aproximado dos cantos da placa
pts_original = np.float32([[389, 220], [618,158],[360,300],[673,250]])

#correção da perspectiva
pts_corrigida = np.float32([[0, 0], [cols , 0], [-100, rows - 10 ], [rows + 800, cols - 500]])

M = cv2.getPerspectiveTransform(pts_original, pts_corrigida)

img_corrigida = cv2.warpPerspective(img_blur_recon, M, (cols, rows))

plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(np.clip(img_blur_recon, 0, 1), cmap='gray')
plt.subplot(122)
plt.imshow(img_corrigida, cmap='gray')
plt.show()
```

Com o tratamento com o filtro feito na aula 6, o resultado ficou muito satisfatório. A imagem ficou mais nítida, permitindo assim, a leitura da placa.


## Scikit-image


Outra forma de simular o filtro de Wiener é utilizar uma função da biblioteca scikit-image. Nesta função basta colocar como argumento a imagem, um kernel e o K. Além disso, nessa função o tamanho do Kernel é adaptado automaticamente, diferente do código acima, onde o kernel deve ter o mesmo tamanho que a imagem.

```python
#Testamos os valores de K, l e ang
K = 0.015
l = 20
ang = -10

h = make_psf((500,500), l, ang)


img_blur_recon_sci = wiener(img_blur, h, K)

plt.figure(figsize=(8, 8))
plt.imshow(np.clip(img_blur_recon_sci, 0, 1), cmap='gray');
```

```python
#redimencionando a janela de exibição da placa
rows, cols = 400, 1000

#achamos os pontos aproximado dos cantos da placa
pts_original = np.float32([[389, 220], [618,158],[360,300],[673,250]])

#correção da perspectiva
pts_corrigida = np.float32([[0, 0], [cols , 0], [-100, rows - 10 ], [rows + 800, cols - 500]])

M = cv2.getPerspectiveTransform(pts_original, pts_corrigida)

img_corrigida_sci = cv2.warpPerspective(img_blur_recon_sci, M, (cols, rows))

plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(np.clip(img_blur_recon_sci, 0, 1), cmap='gray')
plt.subplot(122)
plt.imshow(img_corrigida_sci, cmap='gray')
plt.show()
```

O tratamento utilizando a biblioteca scikit-image, se mostrou razoável. Apesar da placa ter ficado mais legível, a qualidade da imagem ficou inferior comparado com o da aula 6.

## Clahe

A primeira técnica que utilizamos para tentar melhorar o contraste da placa foi a função clahe.

```python
clahe = cv2.createCLAHE(clipLimit=5, tileGridSize=(80,80))
img_clahe = clahe.apply(np.array(img_corrigida*255, dtype=np.uint8))
plt.imshow(img_clahe,cmap='gray');
```

Como foi possível observar, o resultado não ficou muito bom. Apesar de ter melhorado a leitura, a imagem como um todo ficoou com um tom mais claro.

##  Dynamic Range

A segunda forma de melhorar o contraste, é utilizando a técnica de aumentar o dynamic range.

```python
#encontrar os valores máximos e mínimos da imagem
low = img_corrigida.min()
high = img_corrigida.max()
low, high
```

```python
def dr(img, min, max):
    return np.uint8(np.clip((255/(max-min))*(img-min),0,255))
    
new_img = dr(img_corrigida, low, high)

plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(img_corrigida, cmap='gray')
plt.subplot(122)
plt.imshow(new_img, cmap='gray')
plt.show()
```

O resultado final não teve grande diferença comparado a imagem original, como pode ser visto acima. Isso se deve ao fato do dynamic range ser muito grande (low e high, próximos de 0 e 1 respectivamente).

## Threshold 

Por fim, pensamos em utilizar um limiar binário, para assim, aumentar o contraste da imagem tranformando as regiões em totalmente pretas ou brancas (255 ou 0).

```python
limiar, img_dark_square_inverted = cv2.threshold(np.array(img_corrigida*255, dtype=np.uint8), 150, 255, cv2.THRESH_BINARY)
plt.imshow(img_dark_square_inverted, cmap='gray', vmin=0, vmax=255);
```

Esse é o resultado utilizando a imagem original. Ficou bastante legível, apesar de alguns ruídos.

Para ser mais eficientes, pensamos em utilizar a imagem com o tratamento no dynamic range e aplicar a mesma técnica para ver se os ruídos diminuiriam.

```python
limiar, img_dark_square_inverted = cv2.threshold(new_img, 150, 255, cv2.THRESH_BINARY)
plt.imshow(img_dark_square_inverted, cmap='gray', vmin=0, vmax=255);
```

O resutado final foi bastante satisfatório. A imagem ficou bastante legível, inclusive as menores palavras como "Dr 1200".
